```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
df
```

```python
df.isnull().sum()
```

```python
#drop na rows
#drop na - uses default arguements
df.dropna()
```

```python
#drop na's default arguments:
df.dropna(axis="index",how="any")
#axis - index/columns: index- drop na values when rows are missing values,
#columns - drops columns if they had missing values
#how: criteria it uses for dropping
#so, index+how any: drop rows with any missing values 
```

```python
df #none of these are inplace
```

```python
#only drop rows where all values are missing:
df.dropna(axis="index",how="all")
```

```python
#drop if email is NA:
df.dropna(axis="index",how="any",subset=["email"])
```

```python
df.dropna(axis="index",how="any",subset=["email","last"])
```

```python
#email AND last are missing
df.dropna(axis="index",how="all",subset=["email","last"])
```

```python
#handling custom missing values: NA, Missing while loading:
df.replace("NA",np.nan,inplace=True)
df.replace("Missing",np.nan,inplace=True)
df
```

```python
df.dropna(axis="index",how="any",subset=["email","last"])
```

```python
df.isna() #gives mask of values whether or not they are NA
```

```python
#fill NA values with numeric values: eg. missing grades as 0
df.fillna("MISSING")
```

```python
df.fillna(0)
```
